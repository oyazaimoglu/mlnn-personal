```python
#1: Finish creating the following function that takes a list and returns the average value.
def average(my_list):
    total = 0
    for item in my_list:
        total = total + item
    
    avg = total / len(my_list)
    return avg

print("The average is", average([1,2,1,4,3,2,5,9]))
```

```python
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts
```
